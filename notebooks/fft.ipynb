```python
import numpy as np
import matplotlib.pyplot as plt

# create a tensor of increasing sine
x = np.linspace(-1 * np.pi, 1 * np.pi, 1000)
y = 0.2*np.sin(x) + 0.5*np.sin(10*x) + 0.5*np.sin(50*x)


xfft = np.fft.rfft(y)
xfft_mag = np.abs(xfft)

plt.plot(x, y)
plt.show()

# plot the fft with the imaginary part
plt.plot(np.abs(xfft[:100]))
plt.show()
```
