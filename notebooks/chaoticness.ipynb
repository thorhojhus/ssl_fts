```python
import pandas as pd
from hurst import compute_Hc
import numpy as np

def analyze_hurst_exponents(dataset_name):
    # Load the dataset
    df = pd.read_csv(f'../data/{dataset_name}.csv')

    # Replace zero values with NaN to avoid division by zero and assign back to the column
    df['OT'] = df['OT'].replace(0, np.nan)

    # Drop rows where 'OT' is NaN
    df.dropna(subset=['OT'], inplace=True)

    # Compute Hurst exponent for different kinds
    H0, c, data = compute_Hc(df['OT'], kind='change')
    H1, c, data = compute_Hc(df['OT'], kind='random_walk')
    
    try:
        H2, c, data = compute_Hc(df['OT'], kind='price')
        H2_output = f"{H2:.4f} - 'price'       : a series is a cumulative product of changes (i.e. np.cumprod(1+epsilon*np.random.randn(...)))"
    except FloatingPointError:
        H2_output = "'price'       : encountered an error, output omitted"
    
    print(f"""H for OT
{H0:.4f} - 'change'      : a series is just random values (i.e. np.random.randn(...))
{H1:.4f} - 'random_walk' : a series is a cumulative sum of changes (i.e. np.cumsum(np.random.randn(...)))
{H2_output}""")

    return df
```
